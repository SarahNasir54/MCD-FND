```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch

train_csv_path = "/content/drive/MyDrive/Data/instagram/train.csv"
test_csv_path = "/content/drive/MyDrive/Data/instagram/test.csv"

train_df = pd.read_csv(train_csv_path)
test_df = pd.read_csv(test_csv_path)
```

```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""  # or some other default value

    try:
        text = text.encode('utf-8').decode('utf-8').lower()
    except:
        text = text.lower()

    text = re.sub(u"\u2019|\u2018", "\'", text)
    text = re.sub(u"\u201c|\u201d", "\"", text)
    text = re.sub(u"[\u2000-\u206F]", " ", text)
    text = re.sub(u"[\u20A0-\u20CF]", " ", text)
    text = re.sub(u"[\u2100-\u214F]", " ", text)
    text = re.sub(r"http:\ ", "http:", text)
    text = re.sub(r"http[s]?:[^\ ]+", " ", text)
    text = re.sub(r"&gt;", " ", text)
    text = re.sub(r"&lt;", " ", text)
    text = re.sub(r"&quot;", " ", text)
    text = re.sub(r"\"", " ", text)
    text = re.sub(r"#\ ", "#", text)
    text = re.sub(r"\\n", " ", text)
    text = re.sub(r"\\", " ", text)
    text = re.sub(r"[\(\)\[\]\{\}]", r" ", text)
    text = re.sub(u'['
                  u'\U0001F300-\U0001F64F'
                  u'\U0001F680-\U0001F6FF'
                  u'\u2600-\u26FF\u2700-\u27BF]+',
                  r" ", text)
    text = re.sub(r"\'s", " is ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " had ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"#", " #", text)
    text = re.sub(r"\@", " \@", text)
    text = re.sub(r"[\!\?\.\,\+\-\$\%\^\>\<\=\:\;\*\(\)\{\}\[\]\/\~\&\'\|]", " ", text)
    text = text.strip()

    return text
```

```python
def preprocess_dataframe(df):
    df['statements'] = df['statements'].astype(str).apply(clean_text)
    df['comments'] = df['comments'].astype(str).apply(clean_text)
    return df
```

```python
train_df_processed = preprocess_dataframe(train_df)
test_df_processed = preprocess_dataframe(test_df)
```

```python
# Tokenize texts
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Batch processing for feature extraction
def extract_features_batched(texts, model, tokenizer, batch_size=32, max_length=512, device='cuda'):
    model.to(device)
    model.eval()
    features = []

    dataloader = DataLoader(texts, batch_size=batch_size)

    with torch.no_grad():
        for batch in dataloader:
            inputs = tokenizer(batch, padding='max_length', max_length=max_length, truncation=True, return_tensors="pt").to(device)
            outputs = model(**inputs)
            last_hidden_states = outputs.last_hidden_state
            sentence_embeddings = last_hidden_states.mean(dim=1)  # Average pooling
            features.extend(sentence_embeddings.cpu().numpy())

    return features
```

```python
# Extract features for training data
train_statements = train_df_processed['statements'].tolist()
train_comments = train_df_processed['comments'].tolist()
test_statements = test_df_processed['statements'].tolist()
test_comments = test_df_processed['comments'].tolist()
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Extract BERT features
train_statement_features = extract_features_batched(train_statements, model, tokenizer, device=device)
train_comment_features = extract_features_batched(train_comments, model, tokenizer, device=device)
test_statement_features = extract_features_batched(test_statements, model, tokenizer, device=device)
test_comment_features = extract_features_batched(test_comments, model, tokenizer, device=device)
```

```python
# Add features to DataFrames
train_df_processed['statement_features'] = train_statement_features
train_df_processed['comment_features'] = train_comment_features
test_df_processed['statement_features'] = test_statement_features
test_df_processed['comment_features'] = test_comment_features
```

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms
from PIL import Image
import os
```

```python
class ImageDataset(Dataset):
    def __init__(self, image_list):
        self.image_list = image_list
        self.image_ids = list(image_list.keys())

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        image = self.image_list[image_id]
        return image, image_id
```

```python
# Read images and convert to tensor
def read_image(image_path):
    image_list = {}
    data_transforms = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    for filename in os.listdir(image_path):
        try:
            im = Image.open(os.path.join(image_path, filename)).convert('RGB')
            im = data_transforms(im)
            image_list[filename.split('/')[-1].split(".")[0].lower()] = im
        except Exception as e:
            print(f"Error processing image {filename}: {e}")
    print("Image count: " + str(len(image_list)))
    return image_list
```

```python
image_path = "/content/drive/MyDrive/Data/instagram/images/"
image_list = read_image(image_path)
```

```python
dataset = ImageDataset(image_list)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)
```

```python
vgg19 = models.vgg19(pretrained=True).to(device)
vgg19.eval()
```

```python
# Extract features using the trained model
def get_features(image_list, model, device):
    features = {}
    model.classifier = nn.Sequential(*list(model.classifier.children())[:1])  # Keep only the first fully connected layer
    with torch.no_grad():
        for image_id, image_tensor in image_list.items():
            image_tensor = image_tensor.unsqueeze(0).to(device)
            feature = model(image_tensor)
            features[image_id] = feature.cpu()
    return features
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
features = get_features(image_list, vgg19, device)
```

```python
for image_id, feature in features.items():
    print(f"Image ID: {image_id} - Feature Shape: {feature.shape}")
```

```python
# Collect features from the dictionary
image_features_dict = {image_id: feature.squeeze(0) for image_id, feature in features.items()}

train_image_features = []
test_image_features = []
```

```python
for image_id in train_df_processed['imageid']:
    image_id = str(image_id).lower()
    if image_id in image_features_dict:
        train_image_features.append(image_features_dict[image_id])
    else:
        train_image_features.append(torch.zeros(4096))
```

```python
for image_id in test_df_processed['imageid']:
    image_id = str(image_id).lower()
    if image_id in image_features_dict:
        test_image_features.append(image_features_dict[image_id])
    else:
        test_image_features.append(torch.zeros(4096))
```

```python
# Convert lists to tensors
train_image_features_tensor = torch.stack(train_image_features)
test_image_features_tensor = torch.stack(test_image_features)
```

```python
# Convert text features to tensors
train_statement_features_tensor = torch.tensor(train_statement_features, dtype=torch.float32)
train_comment_features_tensor = torch.tensor(train_comment_features, dtype=torch.float32)
test_statement_features_tensor = torch.tensor(test_statement_features, dtype=torch.float32)
test_comment_features_tensor = torch.tensor(test_comment_features, dtype=torch.float32)
```

```python
def prepare_features(statement_features, comment_features, image_features):
    combined_features = torch.cat((statement_features, comment_features, image_features), dim=1)
    return combined_features
```

```python
train_features = prepare_features(train_statement_features_tensor, train_comment_features_tensor, train_image_features_tensor)
test_features = prepare_features(test_statement_features_tensor, test_comment_features_tensor, test_image_features_tensor)
```

```python
print(f"Train statement features shape: {train_statement_features_tensor.shape}")
print(f"Train comment features shape: {train_comment_features_tensor.shape}")
print(f"Test statement features shape: {test_statement_features_tensor.shape}")
print(f"Test comment features shape: {test_comment_features_tensor.shape}")
print(f"Train image features shape: {train_image_features_tensor.shape}")
print(f"Test image features shape: {test_image_features_tensor.shape}")
```

```python
from sklearn.preprocessing import LabelEncoder
train_label = train_df_processed['label'].tolist()
test_label = test_df_processed['label'].tolist()

# Encode labels
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_label)
test_labels_encoded = label_encoder.transform(test_label)

train_labels_tensor = torch.tensor(train_labels_encoded, dtype=torch.long)
test_labels_tensor = torch.tensor(test_labels_encoded, dtype=torch.long)
```

```python
from torch.utils.data import Dataset, TensorDataset
train_dataset = TensorDataset(train_features, train_labels_tensor)
test_dataset = TensorDataset(test_features, test_labels_tensor)

batch_size = 128
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Define the model
class CombinedModel(nn.Module):
    def __init__(self):
        super(CombinedModel, self).__init__()
        self.fc1 = nn.Linear(768 * 2 + 4096, 64)
        self.activation1 = nn.ReLU()
        self.fc2 = nn.Linear(64, 32)
        self.activation2 = nn.ReLU()
        self.fc3 = nn.Linear(32, len(label_encoder.classes_))  # Number of classes 2

    def forward(self, x):
        x = self.fc1(x)
        x = self.activation1(x)
        x = self.fc2(x)
        x = self.activation2(x)
        x = self.fc3(x)
        return x
```

```python
# Model initialization
model = CombinedModel().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
num_epochs = 10
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
# Evaluation function
def evaluate_model(model, dataloader, device):
    model.eval()
    all_labels = []
    all_predictions = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(preds.cpu().numpy())

    accuracy = accuracy_score(all_labels, all_predictions)
    f1 = f1_score(all_labels, all_predictions, average='weighted')
    precision = precision_score(all_labels, all_predictions, average='weighted')
    recall = recall_score(all_labels, all_predictions, average='weighted')

    return accuracy, f1, precision, recall, all_labels, all_predictions
```

```python
train_accuracy, train_f1, train_precision, train_recall, train_labels, train_preds = evaluate_model(model, train_loader, device)
print(f"Training Accuracy: {train_accuracy:.4f}, F1 Score: {train_f1:.4f}, Precision: {train_precision:.4f}, Recall: {train_recall:.4f}")
```

```python
test_accuracy, test_f1, test_precision, test_recall, test_labels, test_preds = evaluate_model(model, test_loader, device)
print(f"Testing Accuracy: {test_accuracy:.4f}, F1 Score: {test_f1:.4f}, Precision: {test_precision:.4f}, Recall: {test_recall:.4f}")
```

```python
from sklearn.metrics import classification_report
train_classification_report = classification_report(train_labels, train_preds)
print("Training Classification Report:\n", train_classification_report)
```

```python
test_classification_report = classification_report(test_labels, test_preds)
print("Testing Classification Report:\n", test_classification_report)
```
